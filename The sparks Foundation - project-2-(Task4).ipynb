```python
df = terrorism_df.copy()
```

```python
df.dropna(axis=1,how='all')
```

```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target',
                  'nkill':'killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                  'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
df=df[['Year','Month','Day','Country','Region','city','latitude','longitude','AttackType','killed','Wounded','Target','Summary','Group','Target_type','Weapon_type',
      'Motive']]
df['casualities']=df['killed']+df['Wounded']
df.head(5)
```

```python
df.isnull().sum()
```

```python
print('Country with Highest Terrorist Attacks:',df['Country'].value_counts().index[0])
print('Region with Highest Terrorist Attacks:',df['Region'].value_counts().index[0])
print('Regions with Highest Terrorist Attacks:',df['Weapon_type'].value_counts().index[0])
print('Maximum people killed in an attack are:',df['killed'].max(),'that took place in',df.loc[df['killed'].idxmax()].Country)
      

```

# Analysis

# Attacking Methods used 

```python
plt.subplots(figsize=(15,6))
sns.countplot('AttackType',data=df,palette='gist_rainbow_r',order=df['AttackType'].value_counts().index)
plt.xticks(rotation=60)
plt.title('Attacking Methods by Terrorists')
plt.show()
```

```python
# Insights:
#     Bombing and Explosion is the most used attacking method No.of people harmed or the caused by this maximum Attack Type
```

# Number Of Terrorist Activities Vs Year

```python
plt.subplots(figsize=(15,6))
sns.countplot('Year',data=df,palette='GnBu',edgecolor=sns.color_palette('dark',7))
plt.xticks(rotation=60)
plt.title('Number of Terrorist Each Year')
plt.show()
```

```python
# Insights:
#     Even measures and actions taken by the government as well by the people terrorist activities are increasing year by year
```

# Who are Targets?

```python
plt.subplots(figsize=(15,6))
sns.countplot(df['Target_type'],palette='rainbow_r',order=df['Target_type'].value_counts().index)
plt.xticks(rotation=65)
plt.title('Favourite Targets')
plt.show()
```

```python
# Insights:
#     Citizens and property is the most targeted.Then it comes the military and the police.Be careful and learn to fight back
```

# Hot zones of Terrorism

```python
plt.subplots(figsize=(15,6))
sns.countplot('Region',data=df,palette='hot',edgecolor=sns.color_palette('dark',7),order=df['Region'].value_counts().index)
plt.xticks(rotation=65)
plt.title('Number of Terrorist Activities By Region')
plt.show()
```

```python
# Insights:
#     Hot Zone is middle east & North Africa,South Asia ranks next.From the graph it can be seen that half of the terrorist activities
#     are conducted in these two area.
```

# Terrorism By Country

```python
plt.subplots(figsize=(18,6))
sns.barplot(df["Country"].value_counts()[:15].index,df['Country'].value_counts()[:15].values,palette='plasma_r')
plt.title('Top Affected Countries')
plt.show()
```

```python
# Insights:
#     Iraq stands first in terms of terrorism and terrorist activities(Based on the dataset from 1970 through 2017).
```

# AtackType Vs Region

```python
pd.crosstab(df.Region,df.AttackType).plot.barh(stacked=True,width=1,color=sns.color_palette('ocean',9))
fig=plt.gcf()
fig.set_size_inches(12,8)
plt.show()
```

```python
# Insights:
#     Various regions are affected by various terriost activites.Observe the region and the attacktype used by the terrorists.
```

# Most Notorious Group

```python
sns.barplot(df['Group'].value_counts()[1:15].values,df['Group'].value_counts()[1:15].index,palette=('CMRmap'))
plt.xticks(rotation=90)
fig=plt.gcf()
fig.set_size_inches(10,8)
plt.title('Terrorist Groups with Highest Terror Attacks')
plt.show()
```

```python
# Insights:
#     Who doesnt know Taliban Interestingly.The one who is not even educated knows that Taliban is the wildest.EDA also concludes
#     the same.Talibanis the most notorious terrorist group.
```

# Attacks vs killed

```python
coun_terror=df['Country'].value_counts()[:15].to_frame()
coun_terror.columns=['Attacks']
coun_kill=df.groupby('Country')['killed'].sum().to_frame()
coun_terror.merge(coun_kill,left_index=True,right_index=True,how='left').plot.bar(width=0.9)
fig=plt.gcf()
fig.set_size_inches(18,6)
plt.show()
```
